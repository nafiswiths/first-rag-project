```python
#create a simple txt file
import os
os.makedirs("../data/text_files", exist_ok=True)
```

```python
sample_text ={ "../data/text_files/sample.txt":"""This is a sample text file.
              A random sample is a subset of individuals chosen from a larger set (population) where every member has an exactly equal chance of being selected.Common Types of Random SamplingSimple Random Sample: Pure chance selection where every individual has an equal probability of being chosen.Stratified Sample: The population is divided into smaller groups (strata), and a random sample is taken from each group.Cluster Sample: The population is divided into existing clusters (like geographic regions), and randomly selected clusters are entirely sampled.Common GeneratorsNumber Generators: You can use tools like the Random.org Integer Generator to select randomized IDs.Text Generators: If you need random filler text for a layout, you can generate it using the Blind Text Generator or using built-in software."""}
for file_path, content in sample_text.items():
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(content)
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("../data/text_files/sample.txt")
loader
documents = loader.load() #to see content
documents
```

directory load


```python
from langchain_community.document_loaders import DirectoryLoader
#load all files of a directory
dir_loader=DirectoryLoader("../data/text_files"
                           , glob="*.txt"#load all txt files/ pattern to match
                           ,loader_cls=TextLoader
                           ,loader_kwargs={"encoding":"utf-8"}
                           ,show_progress=True
                           
                           )
doccuments=dir_loader.load() #to see content
doccuments

```

for pdf 


```python
from langchain_community.document_loaders import PDFLoader
from langchain_community.document_loaders import DirectoryLoader
#load all files of a directory
dir_loader=DirectoryLoader("../data/pdf_files"
                           , glob="*.pdf"#load all pdf files/ pattern to match
                           ,loader_cls=PDFLoader
                           ,loader_kwargs={"encoding":"utf-8"}
                           ,show_progress=True
                           
                           )
pdfs=dir_loader.load() #to see content
pdfs
```
